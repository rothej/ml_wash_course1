```python
import turicreate
```

```python
people = turicreate.SFrame('~/venv/ml_wash/ml_wash_course1/people_wiki/people_wiki.sframe')
```

```python
# Pull Elton John's info
elton = people[people['name'] == 'Elton John']
```

```python
# Most common words in elton table
elton['word_count'] = turicreate.text_analytics.count_words(elton['text'])
elton.stack('word_count',new_column_name=['word','count'])
```

```python
# new table for elton word counts, sort highest first
elton_word_count_table = elton[['word_count']].stack('word_count', new_column_name = ['word','count'])
elton_word_count_table.sort('count',ascending=False)
```

```python
# Calculate TF-IDF for entire corpus
people['word_count'] = turicreate.text_analytics.count_words(people['text'])
people['tfidf'] = turicreate.text_analytics.tf_idf(people['text']) # tf_idf column added
elton = people[people['name'] == 'Elton John'] # replace old table w new TF-IDF counts
elton[['tfidf']].stack('tfidf',new_column_name=['word','tfidf']).sort('tfidf',ascending=False)
```

```python
# use turicreate.distances.cosine to compute cosine distance between Elton and Mccartney
mccartney = people[people['name'] == 'Paul McCartney']
turicreate.distances.cosine(elton['tfidf'][0],mccartney['tfidf'][0])
```

```python
# now calc beckham
beckham = people[people['name'] == 'Victoria Beckham']
turicreate.distances.cosine(elton['tfidf'][0],beckham['tfidf'][0])
```

```python
# Build the two NN models
wnn_model = turicreate.nearest_neighbors.create(people,features=['word_count'],distance='cosine',label='name')
tnn_model = turicreate.nearest_neighbors.create(people,features=['tfidf'],distance='cosine',label='name')
```

```python
# Nearest to Elton using word count
wnn_model.query(elton)
```

```python
# Nearest to Elton using TFIDF
tnn_model.query(elton)
```

```python
# Nearest to Beckham using word count
wnn_model.query(beckham)
```

```python
# Nearest to Beckham using TFIDF
tnn_model.query(beckham)
```
