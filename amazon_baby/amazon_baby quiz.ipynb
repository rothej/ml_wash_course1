```python
import turicreate as tc
```

```python
products = tc.SFrame('~/venv/ml_wash/ml_wash_course1/amazon_baby/amazon_baby.sframe')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
```

```python
products
```

```python
def f_wordcount(i_word):
 if i_word in products['word_count']:   # if it exists in the dictionary
  return products['word_count'][i_word] # apply its value associated w that key
 else:                                  # but if not in dict, count is 0
  return 0
```

```python
for input_val in selected_words:
 products[input_val] = products['word_count'].apply(input_val, f_wordcount)
```

```python
for input_val in selected_words:
    products[input_val] = products['word_count'].apply(lambda x: x[input_val] if input_val in x else 0) 
```
