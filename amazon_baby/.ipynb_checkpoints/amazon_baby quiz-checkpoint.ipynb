```python
import turicreate as tc
```

```python
products = tc.SFrame('~/venv/ml_wash/ml_wash_course1/amazon_baby/amazon_baby.sframe')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
```

# Above fxn did not work, but lambda-izing it did

```python
for input_val in selected_words:
    products[input_val] = products['word_count'].apply(lambda x: x[input_val] if input_val in x else 0) 
```

```python
products
```

```python
for input_val in selected_words:    # cycle thru each column and print sum
 print(str(input_val) + " - " + str(products[input_val].sum()))
```

```python
products['rating'].show() # show ratings of categories
```

```python
# ignore 3* reviews
products = products[products['rating'] != 3]
```

```python
# define positive reviews
products['sentiment'] = products['rating'] >= 4
```

```python
# Train the sentiment classifier
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_words_model = tc.logistic_classifier.create(train_data,target='sentiment', features=['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate'], validation_set=test_data)
```

```python
products['predicted_sentiment'] = selected_words_model.predict(products, output_type = 'probability')
```

```python
selected_coeff = selected_words_model.coefficients.sort('value')
```

```python
selected_coeff
```

```python
selected_words_model.evaluate(test_data, metric='roc_curve')
```
