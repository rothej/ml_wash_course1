```python
# split training sets by label into 4 categories
cat = image_train.filter_by('cat','label')
dog = image_train.filter_by('dog','label')
automobile = image_train.filter_by('automobile','label')
bird = image_train.filter_by('bird','label')
```

# Create a different NN model for each category subset

```python
cat_model = tc.nearest_neighbors.create(cat,
                                        features = ['deep_features'],
                                        label = 'id')
```

```python
dog_model = tc.nearest_neighbors.create(dog,
                                        features = ['deep_features'],
                                        label = 'id')
```

```python
automobile_model = tc.nearest_neighbors.create(automobile,
                                                features = ['deep_features'],
                                                label = 'id')
```

```python
bird_model = tc.nearest_neighbors.create(bird,
                                        features = ['deep_features'],
                                        label = 'id')
```

```python
cat_sample = image_test[0:1]
cat_sample['image'].explore()
# Will now find nearest cat and dog labeled images in the training data that is most similar to this
```
