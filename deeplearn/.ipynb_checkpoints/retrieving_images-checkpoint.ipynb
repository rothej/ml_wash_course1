```python
# split training sets by label
cat = image_data.filter_by('cat','label')
dog = image_data.filter_by('dog','label')
automobile = image_data.filter_by('automobile','label')
bird = image_data.filter_by('bird','label')
```

```python
overall_model =tc.logistic_classifier.create(image_data,
                                        target = 'label',
                                        features = ['image_array'])
```

```python
# placeholder
```

```python
def get_images_from_ids(query_result):
    return image_data.filter_by(query_result['reference_label'],'label')
```

```python
cat_model = get_images_from_ids(overall_model.query(cat))
dog_model = get_images_from_ids(overall_model.query(dog))
automobile_model = get_images_from_ids(overall_model.query(automobile))
bird_model = get_images_from_ids(overall_model.query(bird))
```
